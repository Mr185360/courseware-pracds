## What is Artificial Intelligence?

* weak, def., 
    * an artifical system which has human-like *performance* (or better)
        * on a highly specific task (closed / specific problem)
        
* strong, def., 
    * an artifical system which has human-like *performance* (or better)
        * on unspecific or generic problems (open / general problem)
    
* closed problem
    * all problems are of a similar, specific, pre-solved type
    * problem may be "unsolved" but technique to solving problem is known
    * eg., 3 yo child keeping a door open
        * replace with a rock 
        * improve performance (beat human performance)
        * automate..
    * solving closed problems with tools is very easy 
    * known boundary 
    * a small change to the problem *changes* the solution needed 
    
* open problem <- "real intelligence"
    * problem is unsolved
    * no technique yet developed
    * eg., 3yo child wandering around a room
        * exploring environemnt
        * we can change it
        * developing techniques reporduce changes 
        * eg., light switch
    * *in terms of problem*, unbounded

## What is Machine Learning?
* So how do we do AI?
    * simplified form of statistics
    * uses:
        * simple techniques
        * simple languages, libraries
        * can be taught to A-Level student
* Consider: Linear Regression
    * Historical Data: $(x, y)$ -- (Age, Rating)
        * "Experience" $(x, y)$
    * Regression (relationship):  $Rating = 3*Age + 1$
        * The Model is $(3, 1)$
        * "Learnt", "Trained" via historical $(x, y)$
            * "Learning" $mean(x)$
    * Prediction:
        * $y = 3x + 1$
        * $x = 21$
        * $y = 64/100$
            * "inference" model + data
            * parameters combined with new observation
* Neural Networks 
    * The Big Complicated Approach 
    * Lots of little linear regressions
    * Piece-wise linear regressions
